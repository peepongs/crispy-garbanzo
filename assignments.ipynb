```python
#Assignment 1: Simple if Statment: user to input number and print whether it is positive
num = int(input("Enter a number "))
if num > 0:
  print("entered number is positive")

```

```python
#Assignment 2: If-Else statement
num = int(input("Enter a number "))
if num >= 0:
  print("entered number is positive")
else:
  print("entered number is negative")
```

```python
# Assignment 3: if - elif - else statement
num = int(input("Enter a number "))
if num > 0:
  print("entered number is positive")
elif num < 0:
  print("entered number is negative")
else:
  print("entered number is zero")
```

```python
# Assignment 4: Nested if statement:
num = int(input())
if num > 0:
  if num % 2 == 0:
    print("entered number is positive and even")
  else:
    print("entered number is positive and odd")
elif num < 0:
  print("entered number is negative")
else:
  print("entered number is zero")
```

```python
#Assignment 5: for loop
for i in range(10):
  print(i+1)
```

```python
# Assignment 6: while loop
i = 1
while i!= 11:
  print(i)
  i += 1

```

```python
# Assignment 7: Nested loops

for i in range(5):
  for j in range(5):
    print('*', end = " ")
  print()
```

```python
# Assignment 8:

while True:
  user_input = int(input("input a number "))
  if user_input == 0:
    break
  print(user_input)
print("loop terminated")
```

```python
# Assignment 9: continue Statement
for i in range (10):
  if i == 4:
    continue
  print(i+1)
```

```python
# Assignment 10: pass Statement

def empty_func():
  pass

print(empty_func())
```

```python
# Assignment 11: Combining loops and conditionals

n = int(input("enter a number "))
for i in range(1, n+1):
  if i % 2 == 0:
    print(i)
```

```python
#Assignment 12: Factorial using while

n = int(input())
fact = 1
while n != 0:
  fact = fact*n
  n -= 1
print(fact)
```

```python
# Assignment 13: sum of digits using while
n = input()
i = len(n)
sum = 0
while i!=0:
  sum+=int(n[i-1])
  i -=1
print(sum)
```

```python
# Assignment 14: prime number check using for loop
n = int(input())
count = 0
for i in range(1, n+1):
  if n%i == 0:
    count += 1

if count == 2:
  print("number is prime")
else:
  print("number is not prime")
```

```python
# Assignment 15: Fibonacci sequence
n = int(input())
num = []
for i in range(n):
  if i == 0 or i == 1:
    num.append(i)
  else:
    num.append((num[i-2]+num[i-1]))
print(num)

```
