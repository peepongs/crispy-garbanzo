**Question 11:** Write a Python program to perform arithmetic operations: addition, subtraction, multiplication, and division.

```python
# Your code here
def add(a, b):
  return a + b

def sub(a, b):
  return a - b

def mult(a, b):
  return a * b

def quot(a, b):
  return a / b

print("choose the oper: add, subtract, multiply, or divide")
oper = input()
print(" enter two numbers")
a = int(input())
b = int(input())

if oper == "add":
  print(add(a, b))
elif oper == "subtract":
  print(sub(a, b))

elif oper == "multiply":
  print(mult(a, b))

elif oper == "divide":
  print(quot(a, b))

else:
  print("invalid operation")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input("Enter a Year "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 !=0):
  print("leap year")
else:
  print("not a leap year")
```
